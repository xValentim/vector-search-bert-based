```python
import requests
import json
from pprint import pprint
```

```python
# Testes locais ou em produção

endpoint = "http://localhost:8000/query" # Local endpoint
endpoint = "https://vector-search-bert-based-production.up.railway.app/query"

query = "Nas últimas semanas a economia tem ido bastante mal durante o periodo das eleições, o dolar despencou"

response = requests.get(endpoint, params={"input_content": query})

if response.status_code == 200:
    pprint(response.json())
else:
    pprint(response.text)
```
